```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df_corr = pd.read_csv("./figure4/kg_corr.csv")
df_corr["com"] = df_corr["com"].astype("str")
```

```python
["S" + str(i) for i in range(1, 13)]
```

```python
plt.figure(figsize=(4, 6), dpi = 300)
sns.set_context("talk")
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params, font_scale = 1.3)
g = sns.barplot(data = df_corr,
            y = "com", 
            x = "corr",
            color = "gray")
g.set_ylabel("")
g.set_yticklabels("")
#g.yaxis.set_label_position("right")  # label ("Total bill") on the left
g.yaxis.tick_right()  
g.spines["left"].set_visible(False)
g.spines["right"].set_visible(True)
g.set_xlabel("")
g.set_title("Correlation between aligned weights")
```
