```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ka_pca = pd.read_csv("./figure2/kg_h3a_pca_before.csv")
```

```python
plt.figure(figsize=(9,6), dpi = 300)  
sns.set_theme(style="white", context="talk")  

ax = sns.scatterplot(
    data = ka_pca,
    x = "0", y = "1",
    hue="dataset",        # color by population
    style="dataset",        # different markers for data type
    s = 40, alpha = 0.5,     # bigger points, slight transparency
    edgecolor = "k" # black edges to separate overlapping points
  )
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles = handles, title = "Projects", labels = ["1KGP", "H3Africa"],
         bbox_to_anchor=(1.05, 1), loc="upper left")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
diff_dist_afr = pd.read_csv("./figure2/kg_hs_afr.csv")
diff_dist_eur = pd.read_csv("./figure2/kg_hs_eur.csv")
diff_dist_eas = pd.read_csv("./figure2/kg_hs_eas.csv")
```

```python
fig, ax = plt.subplots(1,3, figsize=(12,4), dpi = 300, sharey = True)  
sns.set_theme(style="white", context="talk", font_scale = 0.8)  

sns.histplot(ax = ax[0],data = diff_dist_afr, x = "diff")
ax[0].set_xlim(-15,15)
ax[0].set_xlabel("Length difference")
ax[0].set_ylabel("Number of STRs")
ax[0].axvline(x = -1, color='gray', linestyle='--', linewidth = 1)
ax[0].axvline(x = 1, color='gray', linestyle='--', linewidth = 1)
ax[0].set_title("AFR")

sns.histplot(ax = ax[1], data = diff_dist_eur, x = "diff")
ax[1].set_xlim(-15,15)
ax[1].set_xlabel("Length difference")
ax[1].set_ylabel("Number of STRs")
ax[1].axvline(x = -1, color='gray', linestyle='--', linewidth = 1)
ax[1].axvline(x = 1, color='gray', linestyle='--', linewidth = 1)
ax[1].set_title("EUR")

sns.histplot(ax = ax[2], data = diff_dist_eas, x = "diff")
ax[2].set_xlim(-15,15)
ax[2].set_xlabel("Length difference")
ax[2].set_ylabel("Number of STRs")
ax[2].axvline(x = -1, color='gray', linestyle='--', linewidth = 1)
ax[2].axvline(x = 1, color='gray', linestyle='--', linewidth = 1)
ax[2].set_title("EAS")
```

```python
ka_tsne = pd.read_csv("./figure2/kg_h3a_20_tsne.csv")
ka_tsne = ka_tsne.query("Superpopulation == 'AFR' | Superpopulation == 'H3Africa' ")
ka_tsne.loc[ka_tsne["dataset"] == "kg", "Population"] = "KG"
```

```python
plt.figure(figsize=(10,6), dpi = 300)  
sns.set_theme(style="white", context="talk")  
palette = sns.color_palette("Set2",8)
new_order = [7, 0, 1, 2, 3, 4, 5, 6]
reordered_palette = [palette[i] for i in new_order]
custom_order = ["KG"] + list(ka_tsne.query("dataset == 'h3a' ")["Population"].unique()) 
custom_palette = dict(zip(custom_order, reordered_palette))

ax = sns.scatterplot(
    data = ka_tsne,
    x = "0", y = "1",
    hue = "Population",        # color by population
    style = "dataset",        # different markers for data type
    s = 40, alpha = 0.5,     # bigger points, slight transparency
    edgecolor = "k",# black edges to separate overlapping points
    palette = custom_palette  # Set3
  )
ax.set_xlabel("tSNE-1")
ax.set_ylabel("tSNE-2")
handles, labels = ax.get_legend_handles_labels()
labels[0] = "Populations"
labels[1] = "AFR (1KGP)"
labels[-3:] = ["Projects", "1KGP", "H3Africa"]
ax.legend(handles = handles, labels = labels,
         bbox_to_anchor=(1.05, 1), loc="upper left")
#ax.legend(title="Population", bbox_to_anchor=(1.05, 1), loc="upper left")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
hs_pca = pd.read_csv("./figure2/hg_sg_pca.csv")
```

```python
plt.figure(figsize=(9,6), dpi = 300)  
sns.set_theme(style="white", context="talk")  

custom_order = ["EUR", "AMR", "AFR", "SAS", "EAS", "CAS", "MIE", "OCE"]
custom_palette = dict(zip(custom_order, sns.color_palette("muted", len(custom_order))))

ax = sns.scatterplot(
    data=hs_pca,
    x = "4", y = "5",
    hue="pop_id",        # color by population
    style="data",        # different markers for data type
    s = 40, alpha = 0.6,     # bigger points, slight transparency
    edgecolor = "k", # black edges to separate overlapping points
    palette= custom_palette )
ax.set_xlabel("PC5")
ax.set_ylabel("PC6")
handles, labels = ax.get_legend_handles_labels()

order = [labels[1:9].index(l) for l in custom_order]
handles[1:9] = [handles[i+1] for i in order]
ax.legend(handles=handles, labels = ["Populations", "EUR", "AMR", "AFR", "SAS", "EAS", "CAS", "MIE", "OCE",
     'Projects','HGDP','SGDP'],
         bbox_to_anchor=(1.05, 1), loc="upper left")
#ax.legend(title="Population", bbox_to_anchor=(1.05, 1), loc="upper left")
sns.despine()
plt.tight_layout()
plt.show()
```
